### 4 Exercise Solutions

```python
jo.value_counts()
```

Which country has won most men's gold medals in single badminton ove the year

```python
mens_badminton = oo[(oo.Gender == 'Men') & (oo.Discipline == 'Badminton') & (oo.Event == 'singles')]
mens_badminton = mens_badminton.sort_values(by=['Athlete'])
mens_badminton
```

```python
mens_badminton.NOC.value_counts()
```

Which three countries have won the most medals in recent years (1984 to 2008)

```python
top_three_countries = oo[(oo.Edition >= 1984) & (oo.Edition <= 2008)]
top_three_countries.NOC.value_counts().head(3)
```

Display the male golde medal winners for the 100 meter track and field sprint event over the years. 
List the result starting with the most recent. 
Show the olympic city, eddition, athlete and the country they represent.

```python
male_gold_medal_winners = oo[(oo.Gender == 'Men') & (oo.Medal == 'Gold') & (oo.Event.str.contains('100m')) & (oo.Discipline == 'Athletics')]
male_gold_medal_winners = male_gold_medal_winners.sort_values(by='Edition', ascending=False)
male_gold_medal_winners[['City', 'Edition', 'Athlete', 'NOC']]
```

## 8. Reshaping

```python
ath_mw = oo[(oo.Edition == 2008) & ((oo.Event == '100m')|(oo.Event == '200m'))]
ath_mw.head()
```

```python
ath_mw_group = ath_mw.groupby(['NOC', 'Gender', 'Discipline', 'Event']).size()
ath_mw_group
```

```python
df = ath_mw_group.unstack(['Discipline', 'Event'])
df
```

```python
df.stack()
```

```python
df.stack('Event')
```

```python
df.unstack()
```

```python
df.unstack('Gender')
```

### 8. Exercise Solutions

Plot the number of gold medals won by USA male and female athletes througout olympics history

```python
umfagm = oo[(oo.NOC == 'USA') & (oo.Medal == 'Gold')]
umfagm.head()
```

```python
umfagm_group = umfagm.groupby(['Edition', 'Gender'])
umfagm_group.size().unstack(fill_value=0).plot(kind='line');
```

Plot five athletes who have won the most gold medals in history of Olympics. When there is a tie consider the number of silver medals and then bronze medals

```python
top5Athlete = oo.groupby(['Athlete', 'Medal']).size().unstack('Medal', fill_value=0)
top5Athlete.sort_values(['Gold', 'Silver', 'Bronze'], ascending=False)[['Gold', 'Silver', 'Bronze']].head(5).plot(kind='bar');
```
